```python
X_train = train_data.imgs
```

```python
X_val = val_data.imgs
```

```python
X_test = test_data.imgs
```

```python
X_train[0]
```

```python
# Normalize values to 0 to 1
X_train = X_train / 255
X_val = X_val / 255
X_test = X_test / 255
```

```python
print(X_train.shape)
print(X_val.shape)
print(X_test.shape)
```

```python
y_train = train_data.labels
y_val = val_data.labels
y_test = test_data.labels
```

```python
y_train[1:10]
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_train_cat = to_categorical(y_train)
y_val_cat = to_categorical(y_val)
y_test_cat = to_categorical(y_test)
```

```python
y_train_cat[1:7]  # binary classification !!!
```

```python
print(y_train_cat.shape)
print(y_val_cat.shape)
print(y_test_cat.shape)
```

```python
y_test_cat[1:7]
```

```python
X_train = X_train.reshape(4708, 64, 64, 1)  # n_channels = 1
X_val = X_val.reshape(524, 64, 64, 1)  # n_channels = 1
X_test = X_test.reshape(624, 64, 64, 1)  # n_channels = 1
```

```python
X_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Dense, Flatten, MaxPool2D, Dropout
```

```python
model = Sequential()

model.add(Conv2D(filters=64, kernel_size=(5,5), input_shape = (64, 64, 1), padding='valid', strides=(1,1), activation='relu'))
model.add(MaxPool2D(pool_size=(2,2), padding='valid'))
model.add(Dropout(0.1))

model.add(Conv2D(filters=64, kernel_size=(4,4), padding='valid', strides=(1,1), activation='relu'))
model.add(MaxPool2D(pool_size=(2,2), padding='valid'))
model.add(Dropout(0.1))

model.add(Flatten()) # 64*64 = 4096

model.add(Dense(units=256, activation='relu')) 
model.add(Dropout(0.5))

model.add(Dense(units=16, activation='relu')) 

model.add(Dense(units=1, activation='sigmoid'))   #binary classification

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
from tensorflow.keras.utils import plot_model
```

```python
plot_model(model=model, show_shapes=True, show_layer_names=True, expand_nested=True, show_layer_activations=True)
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor='val_loss', patience=10)
```

```python
print(X_train.shape)
print(y_train_cat.shape)
print(X_val.shape)
print(y_val_cat.shape)
```

Since this is a Binary Classsification Problem, we might as well use Binary_CrossEntropy as our Loss Function  
for the CNN Model fitting, 
And for Binary_CrossEntropy fit, the output is expected to be (N, 1) and as probabilities rather than (n, no_categories)  
which is for Categorical_CrossEntropy.  
So we do not need to use y_train_cat and y_val_cat for keras Model training using Binary_CrossEntropy.  
Instead we use y_train and y_val for training the model.  

```python
model.fit(x=X_train, y=y_train, validation_data=(X_val, y_val), epochs=50, callbacks=[early_stop])
```

```python
metrics = pd.DataFrame(model.history.history)
```

```python
metrics.columns
```

```python
metrics[['accuracy', 'val_accuracy']].plot()
```

```python
metrics[['loss', 'val_loss']].plot()
```

```python
predicitons = model.predict(X_test)
```

```python
predicitons[1:10]
```

```python
y_test_pred = (predicitons > 0.5).astype(np.int32)
```

```python
y_test_pred[1:8]
```

```python
print(classification_report(y_test, y_test_pred))
```

```python
print(confusion_matrix(y_test, y_test_pred))
```
