```python
import re
```

```python
def get_pwr_trq_value(entry):
    # Return 0 for NaN
    if pd.isna(entry):
        return 0

    # Ensure entry is a string
    s = str(entry)

    # Find the first continuous digits (supports optional leading minus)
    match = re.search(r'-?\d+', s)
    if match:
        return int(match.group())
    else:
        return 0
```

```python
df_cars['Max Power'].apply(get_pwr_trq_value)
```

```python
df_cars['Max Power Value'] = df_cars['Max Power'].apply(get_pwr_trq_value)
```

```python
df_cars['Max Torque Value'] = df_cars['Max Torque'].apply(get_pwr_trq_value)
```

Since the value counts of LPG, Hybrid, Electric, LPG + Electric are all single digits we might also put them in other  

```python
def consolidate_fuel(fuel):
    if fuel == 'Petrol':
        return 'Petrol'
    elif fuel == 'Diesel':
        return 'Diesel'
    elif fuel == 'CNG':
        return 'CNG'
    elif fuel == 'LPG':
        return 'Other'
    elif fuel == 'Electric':
        return 'Other'
    elif fuel == 'CNG + CNG':
        return 'CNG'
    elif fuel == 'Hybrid':
        return 'Other'
    elif fuel == 'Petrol + CNG':
        return 'CNG'
    elif fuel == 'Petrol + LPG':
        return 'Other'
    else:
        return 'Other'
```

```python
df_cars['Fuel Type'].apply(consolidate_fuel).value_counts()
```

```python
df_cars['Fuel'] = df_cars['Fuel Type'].apply(consolidate_fuel)
```

**Model Training and Evaluation**  

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression
```

```python
X = df_cars.drop(labels=['Price'], axis=1, inplace=False)
y = df_cars['Price']
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
LR_model = LinearRegression()
```

```python
LR_model.fit(X_train, y_train)
```

```python
LR_model.intercept_
```

```python
LR_model.coef_
```

```python
y_pred = LR_model.predict(X_test)
```

```python
results = pd.DataFrame({'actual': y_test, 'pred': y_pred})
```

```python
results.columns
```

```python
sns.scatterplot(data=results, x=y_test, y=y_pred)
```

We can see even some negative values and for our target variable like Price, it does not make sense.  
We need to relook at the model .. may be there are some outliers that need to be removed.   

As per Copilot, we need to constrain the linear model to force non negative coeefficients.  

```python
LR_model = LinearRegression(positive=True)
```

```python
LR_model.fit(X_train, y_train)
```

```python
LR_model.intercept_
```

```python
LR_model.coef_
```

```python
y_pred = LR_model.predict(X_test)
```

```python
results = pd.DataFrame({'actual': y_test, 'pred': y_pred})
```

This is definitely better .. atleast we do not see any curve etc... lets evaluate how well this is in terms of metrics etc.   

```python
from sklearn.metrics import mean_absolute_error
```

```python
# Metrics
print(f"R^2:   {r2_score(y_test, y_pred):.3f}")
print(f"MAE:   {mean_absolute_error(y_test, y_pred):.0f}")
print(f"MSE:  {mean_squared_error(y_test, y_pred):.0f}")
print(f'RMSE: {np.sqrt(mean_squared_error(y_test, y_pred)):.0f}')
```
