**Feature Engineering**  

```python
df_titanic_eng = df_titanic.copy()
```

```python
df_titanic_eng.head()
```

```python
df_titanic_eng['Single_Travel'] = (df_titanic_eng['SibSp'] == 0) & (df_titanic_eng['Parch'] == 0)
```

```python
df_titanic_eng.head()
```

```python
def get_last_name(fullname):
    names = str.split(fullname, ',')
    if len(names) > 0:
        return names[0]
    else:
        return ""
```

```python
df_titanic_eng['Last_Name'] = df_titanic_eng['Name'].apply(get_last_name)
```

```python
df_titanic_eng.head()
```

```python
df_titanic['Cabin'].unique()
```

```python
df_titanic_eng['Cabin_Deck'] = df_titanic_eng['Cabin'].apply(lambda x : x[0] if pd.notnull(x) else np.nan)
```

```python
df_titanic_eng.head()
```

```python
sns.countplot(data=df_titanic_eng, x='Cabin_Deck', hue='Survived')
#Belonging to Deck C has a higher percentage of perishing compared to others but since there are many NaN in the cabin data its not confirmed
```

```python
df_titanic_eng.info()
```

**Fill the NA values in Age column**  

```python
# Fliling Empty Age Column by taking average of Age grouped by Sex and Passenger class 
titanic_mean_ages = df_titanic_eng.groupby(['Sex','Pclass'])['Age'].mean()
```

```python
titanic_mean_ages.head()
```

```python
def fill_na_ages(row):
    if pd.isnull(row['Age']):
        return titanic_mean_ages.loc[row['Sex'], row['Pclass']]
    else:
        return row['Age']
```

```python
df_titanic_eng['Age'] = df_titanic_eng.apply(fill_na_ages, axis=1)
```

```python
df_titanic_eng.info()
```

**Fill empty Cabin and Cabin Deck and Embarked**  

```python
df_titanic_eng['Cabin'].fillna("NA", inplace=True)
```

```python
df_titanic_eng['Cabin_Deck'].fillna("NA", inplace=True)
```

```python
df_titanic_eng.head(7)
```

```python
df_titanic_eng['Embarked'].fillna(df_titanic_eng['Embarked'].mode()[0], inplace=True)
```

**Train Test Split**  

```python
from sklearn.model_selection import train_test_split
```

```python
df_titanic_eng['Ticket'].unique()
#Too many unique ticket categories --- can remove the column
```

```python
X = df_titanic_eng.drop(['Survived', 'Name', 'PassengerId', 'Cabin', 'Ticket', 'Last_Name'], axis=1)
#Too many unique ticket categories --- can remove the column
#Similarly can remove the Cabin column -- we can use the cabin_deck instead which has only 7-8 categories
# Name and Passenger Ids are also unique and so need to be removed as they are neither numbers neither categories
# Similarly drop the Survied from X
```

```python
y = df_titanic_eng['Survived']
```

```python
X.head()
```

```python
y.head()
```

```python
X = pd.get_dummies(X, columns=['Sex', 'Embarked', 'Single_Travel', 'Cabin_Deck'], drop_first=True, dtype=int)
```

```python
X.head()
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3, random_state=100)
```

**Model Training using Logistic Regression, Decision Tree and SVM**  
**Hyper Parameter tuning using GridSearchCV**  

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
models = {
    'LogisticRegression': LogisticRegression(),
    'DecisionTree': DecisionTreeClassifier(),
    'SVM': SVC()
}
```

```python
param_info = {
    'LogisticRegression' :
    {
        'C': [0.01, 0.1, 1, 10, 100],         # Regularization strength
        'penalty': ['l2'],                    # L2 regularization (most common)
        'solver': ['liblinear'],             # Works well with small datasets and supports L2
        'max_iter': [100, 500, 1000]         # Ensure convergence
    },
    'DecisionTree':
    {
        'max_depth': [3, 5, 10, None],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'criterion': ['gini', 'entropy']
    },
    'SVM':
    {
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf'],
        'gamma': ['scale', 'auto']
    }
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix
```

```python
best_models = []
for model_name, model in models.items():
    grid = GridSearchCV(model, param_info[model_name], cv=5, scoring='accuracy')
    grid.fit(X_train, y_train)
    print("_"*60)
    print(f'Best Params for model {model_name} are : {grid.best_params_}')
    best_models.append(grid.best_estimator_)
    y_pred = grid.best_estimator_.predict(X_val)
    print(f'Classification Report for model {model_name} is below :')
    print(classification_report(y_val, y_pred))
    print(f'Confusion Matrix for model {model_name} is below : ')
    print(confusion_matrix(y_val, y_pred))
    print("_"*60)
```

**Apply Bagging (Bootstrap Aggregating) to the estimators with best params**  

```python
from sklearn.ensemble import BaggingClassifier
```

```python
best_models
```

```python
#BAgging Classifiers
bagging_models = {
    'Bagging_LogisticRegression' : BaggingClassifier(estimator=best_models[0], n_estimators=100, random_state=100),
    'Bagging_DecitionTree' : BaggingClassifier(estimator=best_models[1], n_estimators=100, random_state=100),
    'Bagging_SVM' : BaggingClassifier(estimator=best_models[2], n_estimators=100, random_state=100)
}
```

```python
for model_name, bag_model in bagging_models.items():
    print(f'\nTraining Model: {model_name} -> {bag_model}')
    print('-' * 60)
    try:
        # Fit
        bag_model.fit(X_train, y_train)

        # Basic sanity checks post-fit
        if hasattr(bag_model, "estimators_"):
            print(f'Number of base estimators trained: {len(bag_model.estimators_)}')
        if hasattr(bag_model, "n_features_in_"):
            print(f'Features seen during fit: {bag_model.n_features_in_}')

        # Predict on validation set
        y_pred = bag_model.predict(X_val)

        # Metrics
        print('Classification Report:')
        print(classification_report(y_val, y_pred))
        print('Confusion Matrix:')
        print(confusion_matrix(y_val, y_pred))

    except Exception as e:
        print(f' Error training {model_name}: {e}')

    print('-' * 60)
```

**After doing GridSearchCV and finding the best hyperparameters and then again using Bagging Classifiers for the best estimators of type
Logistic Regression, DecisionTreeClassifier and SVM classifiers.**  
**The accuracy for the logistic regression bagging classifier was about 80%, for Decision Tree Bagging classifier was 81% and for SVM bagging classifier was 78%.**
